```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

# LABEL ENCODER

```python
df=pd.DataFrame({'Color':['Red','Blue','Green','Red','Yellow']})
```

```python
df
```

```python
label=LabelEncoder()
```

```python
label.fit_transform(df['Color'])
```

```python
label.transform(['Red']) #[2]
label.transform(['Black']) # ValueError 
label.transform([['yes','no']]) #y should be 1D array
```

# ONEHOT ENCODER

```python
df1=pd.DataFrame({'Color':['Red','Blue','Green','Red','Yellow'],'City':['Bengaluru','chennai','hydrabad','Bengaluru','hydrabad']})
df1
```

```python
one_label=OneHotEncoder(handle_unknown='ignore') #sparse_output=False  ---> directly we get the values instead of sparse matrix
                          # dtype='int' --> used convert the values to int default its float
```

```python
nes=one_label.fit_transform(df1[['Color','City']]) #returns sparse matrix 
```

```python
val=nes.toarray() # to extract that values...from the sparse matrix
val
```

```python
res=one_label.get_feature_names_out() # array of labels or categorical values 
res
```

```python
pd.DataFrame(val,columns=res)
```

```python
mat=one_label.transform([['Blue','Bengaluru']])
```

```python
val=mat.toarray()
val
```
