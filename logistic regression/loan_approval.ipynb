```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r'C:\Users\pavan\Desktop\data science\logistic regression\loan_approval_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
x=df.drop(columns=['Loan_Approved'])
y=df['Loan_Approved']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
num_col=x_train.select_dtypes(include='number')
cat_col=x_train.select_dtypes(exclude='number')
```

```python
encoder=OneHotEncoder(sparse_output=False)
```

```python
val=encoder.fit_transform(cat_col)
```

```python
lab=encoder.get_feature_names_out()
```

```python
encoded_col=pd.DataFrame(val,columns=lab)
```

```python
encoded_col
```

```python
new_df=pd.concat([encoded_col.reset_index(),num_col.reset_index()],axis=1)
```

```python
model=LogisticRegression()
```

```python
model.fit(new_df,y_train)
```

# Relation

```python
label=LabelEncoder()
y_train_coded=label.fit_transform(y_train)
y_train_coded_df=pd.DataFrame(y_train_coded,columns=['target'])
```

```python
train_data=pd.concat([new_df.reset_index(drop=True),y_train_coded_df.reset_index(drop=True)],axis=1)
train_data=train_data.drop(columns='index')
```

```python
train_data.corr()['target']
```

# training

```python
print('train score :', model.score(new_df,y_train))
```

# Testing

```python
num_test=x_test.select_dtypes(include='number')
cat_test=x_test.select_dtypes(exclude='number')
```

```python
val=encoder.transform(cat_test)
lab=encoder.get_feature_names_out()
```

```python
df1=pd.DataFrame(val,columns=lab)
test_data=pd.concat([df1.reset_index(),num_test.reset_index()],axis=1)
```

```python
model.predict(test_data)
```

```python
print('test score : ',model.score(test_data,y_test))
```
