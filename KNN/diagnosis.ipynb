```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder,RobustScaler,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv(r'C:\Users\pavan\Desktop\data science\KNN\column_2C_weka.csv')
```

```python
x=df.drop(columns='class')
y=df['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
num=x_train.select_dtypes(include='number').columns
```

```python
num_pipleline=Pipeline(
    steps=[
        ('num_scaling',RobustScaler())
    ]
)


cat_pipleline=Pipeline(
    steps=[
        ('car_encoding',LabelEncoder())
    ]
)
```

```python
preprocessing=ColumnTransformer(
    transformers=[
        ('num_pipleline',num_pipleline,num),
        #('cat_pipleline',cat_pipleline,cat)
    ]
)
```

```python
main_pipeline=Pipeline(
    steps=[
        ("preprocessing",preprocessing),
        ('model',KNeighborsClassifier(metric='manhattan',n_neighbors=5))
    ]
)
```

```python
main_pipeline.fit(x_train,y_train)
```

```python
main_pipeline.score(x_train,y_train)
```

```python
main_pipeline.score(x_test,y_test)
```
