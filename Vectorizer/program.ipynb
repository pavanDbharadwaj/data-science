```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import pandas as pd
```

```python
df=pd.DataFrame(['ML is subset of AI ',
    'AI is superset of both ML and DL',
    'GenAI is used to create or generate any type content'],columns=['col1'])
df
```

```python
cv = CountVectorizer()

X = cv.fit_transform(df['col1'])

df_cv = pd.DataFrame(
    X.toarray(),
    columns=cv.get_feature_names_out()
)

print(df_cv)
```

```python
cv.vocabulary_
```

```python
cv = CountVectorizer(stop_words='english')# max_df,min_df,binary

X = cv.fit_transform(df)

df_cv = pd.DataFrame(
    X.toarray(),
    columns=cv.get_feature_names_out()
)

print(df_cv)
```

```python
t=TfidfVectorizer()
Y = t.fit_transform(df['col1'])

df_t = pd.DataFrame(
    Y.toarray(),
    columns=t.get_feature_names_out()
)

print(df_t)
```

```python
t=TfidfVectorizer(min_df=2)
Y = t.fit_transform(df['col1'])

df_t = pd.DataFrame(
    Y.toarray(),
    columns=t.get_feature_names_out()
)

print(df_t)
```
