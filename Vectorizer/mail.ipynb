```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report,f1_score
```

```python
df=pd.read_csv(r'C:\Users\pavan\Desktop\data science\Vectorizer\mail_data.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
x=df.drop(columns='Category')
y=df['Category']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
cat_pipleline=Pipeline(
    steps=[
        ('vectorizing',CountVectorizer(stop_words='english'))
    ]
)
```

```python
preprocessing=ColumnTransformer(
    transformers=[
        ('cat_pipleline',cat_pipleline,'Message')
    ]
)
```

```python
main_pipeline=Pipeline(
    steps=[
        ("preprocessing",preprocessing),
        ('model',SVC(class_weight='balanced'))
    ]
)
```

```python
main_pipeline.fit(x_train,y_train)
```

```python
main_pipeline.score(x_train,y_train)
```

```python
main_pipeline.score(x_test,y_test)
```

```python
y_pred = main_pipeline.predict(x_test)

f1 = f1_score(y_test, y_pred, average='macro')
print("F1 score:", f1)
```

```python
print(classification_report(y_test, y_pred))
```
