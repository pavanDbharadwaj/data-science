```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,OneHotEncoder,OrdinalEncoder,LabelEncoder,RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor

```

```python
df=pd.read_csv(r'C:\Users\pavan\Desktop\data science\decision tree\ensemble methods\train.csv')
```

```python
df.shape
```

```python
df=df.drop(columns='Open Date')
```

```python
x=df.drop(columns='revenue')
y=df['revenue']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
num=x_train.select_dtypes(include='number').columns
cat=x_train.select_dtypes(exclude='number').columns
cat=cat.drop(labels='City Group')
```

```python
num_pipeline=Pipeline(
    steps=[
        ('num_col',StandardScaler())
    ]
)


cat_pipeline=Pipeline(
    steps=[
        ('cat_col',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
    ]
)

city_pipeline=Pipeline(
    steps=[
        ('city_encodeing',OrdinalEncoder(categories=[['Other', 'Big Cities']],handle_unknown='use_encoded_value',unknown_value=-1))
    ]
)
```

```python
preprocessing=ColumnTransformer(
    transformers=[
        ('num_pipeline',num_pipeline,num),
        ('city_group',city_pipeline,['City Group']),
        ('cat_pipeline',cat_pipeline,cat)
    ]
    
)
```

```python
main_pipeline=Pipeline(
    steps=[
        ('preprocessing',preprocessing),
        ('model',BaggingRegressor())
    ]
)
```

```python
main_pipeline.fit(x_train,y_train)
```
